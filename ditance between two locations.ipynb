```python
import pandas as pd
```

# distance bewteen two locations

```python
from math import sin, cos, sqrt, atan2, radians
```

```python
sub_market=pd.read_csv('submarket_3.csv')
top_msa=pd.read_csv('msa_top.csv')
```

```python
sub_market['tep']=1
top_msa['tep']=1
df= pd.merge(sub_market,top_msa, on=['tep'])
df.drop('tep', axis=1)
```

```python
df['dist']=1
```

```python
R = 6373.0
for i in range(len(df)):
    lat1 = radians(df['lat1'][i])
    lon1 = radians(df['lng1'][i])
    lat2 = radians(df['lat2'][i])
    lon2 = radians(df['lng2'][i])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    df['dist'][i] = R * c
```

```python
df2=df.sort_values(['MSA','dist']).drop_duplicates(['submarket_city'],keep='first')
```

```python
df2.to_csv('dist_submarket_msa_top.csv')
```
