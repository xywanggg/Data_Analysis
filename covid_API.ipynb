# current data

```python
url="https://api.covidtracking.com/v1/states/current.json"
f=urllib.request.urlopen(url)
data=f.read().decode("utf-8")
readfile=json.loads(data)
```

```python
with open('covid.csv','w',newline='') as coviddata:
    csvwriter=csv.writer(coviddata)
    count=0
    for result in readfile:
        if count==0:
            header=result.keys()
            csvwriter.writerow(header)
            count+=1
        csvwriter.writerow(result.values())
```

## history data-covid

```python
covid['date2']=pd.to_datetime(covid['date'], format='%Y%m%d')
```

```python
covid['year'] = covid['date2'].apply(lambda x: x.year)
covid['month'] = covid['date2'].apply(lambda x: x.month)
covid['day'] = covid['date2'].apply(lambda x: x.day)
```

```python
covid=covid.groupby(['state','year','month']).head(1)
```

```python
covid=covid.drop(['date', 'date2','day'], axis=1)
```

```python
covid2=covid[['state','year','month','positive','negative','hospitalized','recovered','death',]]
```

## join

```python
covid2.shape
```

```python
sample.shape
```

```python
c=sample.merge(covid2, left_on=['state','Year','Month'], right_on=['state','year','month'], how='left') #left
```

```python
c.shape
```

```python
c.tail()
```

```python
c=c.drop(['month','year'],axis=1)
```

```python
c.columns
```

```python
c.shape
```

```python
data=c.sort_values(['state','city','Year','Month'])
```

```python
data=data.fillna(0)
```

```python
## create new columns
data['new_positive']=data['positive'].diff(1)
data['new_negative']=data['negative'].diff(1)
data['new_hospitalized']=data['hospitalized'].diff(1)
data['new_recovered']=data['recovered'].diff(1)
data['new_death']=data['death'].diff(1)
```

```python
data=data.fillna(0)
```

```python
#set the new columns which is in next month = 0
now=pd.to_datetime('now')
a=now.month+1
b=now.year
```

```python
col=['new_positive','new_negative','new_hospitalized','new_recovered','new_death']
```

```python
for i in col:
    data[i][(data['Year']==b) & (data['Month']==a)]=0
```
